# Modelo de regresión lineal
## Importación de librerías
Se cargan las librerías necesarias: `numpy` para arreglos, `matplotlib` para graficar y de `scikit-learn` el modelo de regresión lineal junto con métricas de evaluación.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Definición de variables
Se crean los datos de entrada. `X` es la variable independiente en formato 2D y `y` es la variable dependiente con los valores reales.



```python
#y = m*x+b
#Variable independiente
#ScKit-Learn requiere que se usen arreglos de 2D
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
print(X)

#Varialble dependiente
y=np.array([2, 4, 5, 4, 5])
print(y)
```

## Visualización inicial
Se grafica un diagrama de dispersión para observar cómo se relacionan los datos originales entre `X` y `y`.

```python
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Datos originales")
plt.grid(True)
plt.show()
```

## Creación y entrenamiento del modelo
Se construye el modelo de regresión lineal, se entrena con los datos y se obtienen los parámetros de la recta: pendiente y intercepto.


```python
#Generamos nuestro modelo de regresion linear
model= LinearRegression()
#alimentamos nuestro modelo

model.fit(X,y)

#parametros:m(peso), b(sesgo)

m=model.coef_[0]
b = model.intercept_

print(f"Pendiente (m):{m:2f}")
print(f"Intercepto(b):{b:2f}")
```

## Predicciones y métricas
Se generan las predicciones del modelo y se evalúan con el error cuadrático medio (MSE) y el coeficiente de determinación (R²).

```python
#generamos una predicción
y_pred = model.predict(X)
#Evaluamos esa predicción
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y,y_pred)

print(f"Mean Squared Error:{mse:2f}")
print(f"R2 Score:{r2:2f}")
```

## Gráfica final del modelo
Se muestran los datos reales junto con la línea roja del modelo lineal, visualizando cómo la regresión aproxima la relación entre `X` y `y`.

```python
plt.scatter(X,y, label="Datos reales")
plt.plot(X,y_pred,color="red", label="Modelo lineal")
plt.xlabel("X")
plt.ylabel("y")
plt.scatter(X,y)
plt.title("Regresión linear simple")
plt.grid(True)
plt.show()
```

#Implementación de Pipeline

Un pipeline encapsula todo el flujo de procesamiento de datos en un solo objeto.

Para el usuario, entrenar un pipeline es igual que entrenar un modelo.

El escalado no cambia el resultado conceptual, pero prepara el terreno para modelos más complejos.

##Justificación

Hasta ahora hemos entrenado modelos paso por paso.
En la vida real, esos pasos siempre se repiten en el mismo orden. Un pipeline es una forma de decirle a las computadora: **Estos son los pasos, siempre ejecútalos así**.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
def plot_model(xlabel: str, ylabel: str, title: str):
  """
  Función para graficar el modelo de regresión lineal lineal.
  """
  plt.scatter(X,y, label="Datos reales")
  plt.plot(X,y_pred,color="red", label="Modelo lineal")
  plt.xlabel("Distancia al centro")
  plt.ylabel("Costo")
  plt.scatter(X,y)
  plt.title("Estimación de precio de una casa")
  plt.grid(True)
  plt.show()
```

```python
# 1. Definiremos las variables dependientes e independientes de nuestro modelo.

#y = m*x+b
#Variable independiente x distancia al centro (km)
#ScKit-Learn requiere que se usen arreglos de 2D
X = np.arange(1, 21).reshape(-1,1)
print(X)

#Varialble dependiente y: precio (miles de $)

y = np.array([340, 320, 290, 260, 230,
              210, 220, 245, 275, 305,
              330, 350, 365, 375, 380,
              385, 390, 395, 398, 400])
print(y)
```

```python
pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("regresor", LinearRegression())
])

#Entrenamos al modelo.
pipeline.fit(X,y)

#Predicción con el modelo
y_pred = pipeline.predict(X)

plot_model(xlabel="Distancia al centro (km)", ylabel="Costo(Miles de $)",
           title="Estimación del precio de una casa")
```
